```python
# PhD Research Timeline: 2026 Gantt Chart

# This notebook visualizes the research timeline from January to December 2026, accounting for:
# - Research Assistant role (February - May 2026) - PhD work paused
# - Continued work on polyp segmentation explainability
# - Vision-Language Model (VLM) explainability research
# - Publication milestones for early work
# - Evaluation experiments and human studies
```

```python
# Install required libraries if not already installed
# Uncomment the line below if you need to install plotly
# !pip install plotly pandas

import plotly.figure_factory as ff
import pandas as pd
from datetime import datetime, timedelta
import plotly.graph_objects as go

print("✓ Libraries imported successfully")
```

```python
# Define research tasks and timeline (condensed)
tasks = [
    # === JANUARY 2026 ===
    dict(Task="Early segmentation work completion", Start='2026-01-01', Finish='2026-01-31', 
         Category='Segmentation', Status='Active'),
    
    # === RESEARCH ASSISTANT ROLE (PhD PAUSED) ===
    dict(Task="Research Assistant Role (PhD Paused)", Start='2026-02-01', Finish='2026-05-31', 
         Category='Administrative', Status='Paused'),
    
    # === JUNE - AUGUST: RESUME PhD + FIRST PAPER ===
    dict(Task="Resume PhD & first paper preparation", Start='2026-06-01', Finish='2026-08-15', 
         Category='Publication', Status='Planned'),
    dict(Task="🎯 First paper submission (Segmentation)", Start='2026-08-10', Finish='2026-08-15', 
         Category='Milestone', Status='Target'),
    
    # === AUGUST - SEPTEMBER: VLM EXPERIMENTS ===
    dict(Task="VLM explainability experiments (Phase 2)", Start='2026-08-15', Finish='2026-09-30', 
         Category='VLM', Status='Planned'),
    
    # === SEPTEMBER - NOVEMBER: HUMAN STUDIES ===
    dict(Task="Human calibration studies", Start='2026-09-15', Finish='2026-11-30', 
         Category='Human Study', Status='Planned'),
    
    # === OCTOBER - DECEMBER: SECOND PAPER ===
    dict(Task="VLM paper writing & preparation", Start='2026-10-01', Finish='2026-12-15', 
         Category='Publication', Status='Planned'),
    dict(Task="🎯 Second paper submission (VLM)", Start='2026-12-15', Finish='2026-12-20', 
         Category='Milestone', Status='Target'),
]

# Convert to DataFrame
df = pd.DataFrame(tasks)

# Display summary
print(f"Total tasks planned: {len(df)}")
print(f"\nTasks by category:")
print(df['Category'].value_counts())
print(f"\nTimeline: {df['Start'].min()} to {df['Finish'].max()}")
```

```python
# Define color scheme for different categories
colors = {
    'Administrative': 'rgb(200, 200, 200)',  # Grey - RA role pause
    'Admin': 'rgb(150, 150, 150)',  # Dark grey
    'Segmentation': 'rgb(100, 149, 237)',  # Cornflower blue
    'VLM': 'rgb(255, 140, 0)',  # Dark orange
    'Human Study': 'rgb(34, 139, 34)',  # Forest green
    'Publication': 'rgb(220, 20, 60)',  # Crimson red
    'Milestone': 'rgb(255, 215, 0)',  # Gold
}

# Map colors to tasks
df['Color'] = df['Category'].map(colors)

print("✓ Color scheme defined")
print("\nCategory colors:")
for cat, color in colors.items():
    print(f"  {cat}: {color}")
```

```python
# Create Gantt chart using plotly
fig = ff.create_gantt(
    df,
    colors=colors,
    index_col='Category',
    show_colorbar=True,
    group_tasks=True,
    showgrid_x=True,
    showgrid_y=True,
    title='PhD Research Timeline: January - December 2026 (High-Level Overview)',
    bar_width=0.5,
    height=600,
    width=1400
)

# Update layout for better readability
fig.update_layout(
    xaxis=dict(
        title='Timeline',
        tickformat='%b %Y',
        tickmode='linear',
        dtick='M1',  # Monthly ticks
        tickangle=-45,
        gridcolor='lightgray',
        showgrid=True
    ),
    yaxis=dict(
        title='Research Tasks',
        autorange='reversed',
        gridcolor='lightgray',
        showgrid=True
    ),
    font=dict(size=10),
    plot_bgcolor='white',
    paper_bgcolor='white',
    hovermode='closest',
    title_font_size=18,
    title_x=0.5,
    margin=dict(l=350, r=50, t=100, b=100)
)

# Add annotations for key periods
fig.add_annotation(
    x='2026-03-15',
    y=0.95,
    xref='x',
    yref='paper',
    text='<b>PhD PAUSED (RA Role)</b>',
    showarrow=False,
    bgcolor='rgba(200, 200, 200, 0.5)',
    bordercolor='red',
    borderwidth=2,
    font=dict(size=12, color='red')
)

fig.add_annotation(
    x='2026-08-15',
    y=0.95,
    xref='x',
    yref='paper',
    text='<b>PhD ACTIVE: Phase 2 Experiments</b>',
    showarrow=False,
    bgcolor='rgba(255, 140, 0, 0.3)',
    bordercolor='orange',
    borderwidth=2,
    font=dict(size=12, color='darkorange')
)

fig.add_annotation(
    x='2026-11-01',
    y=0.95,
    xref='x',
    yref='paper',
    text='<b>Human Studies & Writing</b>',
    showarrow=False,
    bgcolor='rgba(34, 139, 34, 0.3)',
    bordercolor='green',
    borderwidth=2,
    font=dict(size=12, color='darkgreen')
)

# Show the chart
fig.show()

print("\n✓ Gantt chart created successfully!")
print("\nKey milestones:")
print("  • Jan 2026: Early segmentation work completion")
print("  • Feb-May 2026: Research Assistant role (PhD paused)")
print("  • Jun-Aug 2026: PhD resumes, work towards first paper submission")
print("  • Aug-Sep 2026: VLM explainability experiments (Phase 2)")
print("  • Sep-Nov 2026: Human calibration studies")
print("  • Dec 2026: Second paper submission target")
```

```python
# Create monthly workload visualization
import plotly.express as px

# Calculate tasks per month per category
df['Start'] = pd.to_datetime(df['Start'])
df['Finish'] = pd.to_datetime(df['Finish'])

# Generate monthly breakdown
monthly_data = []
months = pd.date_range('2026-01-01', '2026-12-31', freq='MS')

for month in months:
    month_end = month + pd.offsets.MonthEnd(0)
    for _, task in df.iterrows():
        if task['Start'] <= month_end and task['Finish'] >= month:
            monthly_data.append({
                'Month': month.strftime('%b %Y'),
                'Category': task['Category'],
                'Task': task['Task']
            })

monthly_df = pd.DataFrame(monthly_data)
workload_summary = monthly_df.groupby(['Month', 'Category']).size().reset_index(name='Task Count')

# Create stacked bar chart
fig2 = px.bar(
    workload_summary,
    x='Month',
    y='Task Count',
    color='Category',
    title='Monthly Workload Distribution by Category',
    color_discrete_map=colors,
    height=500,
    width=1200
)

fig2.update_layout(
    xaxis_title='Month',
    yaxis_title='Number of Active Tasks',
    legend_title='Category',
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(size=11),
    title_x=0.5,
    title_font_size=16
)

fig2.show()

print("✓ Monthly workload visualization created")
```
